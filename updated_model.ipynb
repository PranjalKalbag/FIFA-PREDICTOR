```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("FullData.csv")
dataset = dataset.drop(['Nationality','National_Position','National_Kit','Club','Club_Position','Club_Kit','Contract_Expiry','Birth_Date','Work_Rate','Height','Weight',"Club_Joining"], axis = 1)
n = 20
squad = ['GK','LB','CB','CB','RB','LCM','CM','RCM','LW','ST','RW']
my_team=[]
subs=[]
flag = 0
```

```python
pos = ["GK"]
gk = dataset.loc[dataset['Preffered_Position'].isin(pos)] 
gk_names = gk['Name']
gk_crit = gk.iloc[:,[2,3,36,37,38,39,40]].values
gk_rate = gk['Rating']
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
gk_crit[:,0] = le.fit_transform(gk_crit[:,0])
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])],  
    remainder='passthrough'                                        
    )
gk_crit = ct.fit_transform(gk_crit)

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = n)
regressor.fit(gk_crit, gk_rate)

# Predicting a new result
gk_pred = regressor.predict(gk_crit)

gk_compare = list(zip(gk_pred, gk_names,gk_rate))
gk_compare.sort(reverse=True)

my_team.append(gk_compare[0][1])
subs.append(gk_compare[1][1])
```
